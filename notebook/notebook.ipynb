```python
percent_missing = df.isnull().sum() * 100 / len(df)
percent_missing
```

```python
cols_to_drop = percent_missing[percent_missing >10].index
df.drop(columns=cols_to_drop, inplace=True) 
```

```python
print("Dropping columns with >10% missing values:", list(cols_to_drop))
```

```python
# Drop rows where target variables are missing
df = df.dropna(subset=['RainTomorrow'])
```

```python
df.shape
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.Date.dtype
```

```python
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
```

```python
df.drop("Date", axis=1, inplace=True)
```

```python
df.shape
```

```python
df.columns
```

```python
categorical = []
numerical = []

for col in df.columns:
    if df[col].dtype=='object':
        categorical.append(col)
    else:
        numerical.append(col)
```

```python
for col in df.columns:
    if col in numerical:
        df[col]= df[col].fillna(df[col].median())
    else:
        df[col] = df[col].fillna(df[col].mode()[0])
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
for col in categorical:
    label_encoder = LabelEncoder()
    df[col] = label_encoder.fit_transform(df[col])
    label_mapping = dict(zip(label_encoder.classes_,range(len(label_encoder.classes_))))
    print(label_mapping)
```

RainToday- Strong temporal link: If it rained today, there's a higher chance it might rain tomorrow due to weather patterns, so its a input feature.


```python
X = df.drop('RainTomorrow', axis=1)
y = df.RainTomorrow
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.2, stratify=y, random_state=90)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(random_state=42)
model.fit(X_train,y_train)
```

```python
feature_importance_df = pd.DataFrame({'Feature': X_train.columns, 
                                'importance':model.feature_importances_ })

feat_imp_df = feature_importance_df.sort_values('importance', ascending=False)
feat_imp_df['cum_imp'] = np.cumsum(feat_imp_df['importance'])
```

```python
feat_imp_df
```

```python
from sklearn.metrics import accuracy_score, precision_score, f1_score,recall_score
```

```python
y_pred= model.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
f1_score(y_test,y_pred)
```

```python
from sklearn.model_selection import GridSearchCV
scale_pos_weight=neg/pos

param_grid = {
    'n_estimators': [100,200,500,1000],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.001, 0.01, 0.1, 0.2],
    'n_estimators': [100, 300],
    'subsample': [0.8, 1.0],
    'scale_pos_weight': [scale_pos_weight]
}

grid = GridSearchCV(xgb.XGBClassifier(), param_grid, scoring='roc_auc', cv=2)
grid.fit(X_train, y_train)
print("Best params:", grid.best_params_)
```

```python
print("Best params:", grid.best_params_)
```

```python
y_grid_pred = grid.predict(X_test)
print("Accuracy : ", accuracy_score(y_test, y_grid_pred))
print("F1-score : ", f1_score(y_test, y_grid_pred))
```

```python
from sklearn.ensemble import GradientBoostingClassifier, AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
```

```python

classifiers ={
    "Random_Forest": RandomForestClassifier(random_state=42),
    "Gradient Boosting" : GradientBoostingClassifier(random_state=42),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "XGBoost" : XGBClassifier(random_state=42),
    "AdaBoostClassifier" : AdaBoostClassifier(random_state=42),
    "lightGBM" : LGBMClassifier(random_state=42)
    }
```

```python
metrics ={
    "Model" :[],
    "Accuracy":[],
    "Precision": [],
    "Recall": [],
    "F1 score": []
}
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
for model_name, classifier in classifiers.items():
    classifier.fit(X_train,y_train)

    y_pred = classifier.predict(X_test)

    accuracy = accuracy_score(y_test,y_pred)
    precision = precision_score(y_test,y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test,y_pred)

    metrics["Model"].append(model_name)
    metrics["Accuracy"].append(accuracy)
    metrics["Precision"].append(precision)
    metrics["Recall"].append(recall)
    metrics["F1 score"].append(f1)
```

```python
metrics_df= pd.DataFrame(metrics)
metrics_df
```
